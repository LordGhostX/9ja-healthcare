# Folium Lagos Demo

```python
import folium
import pandas as pd
```

```python
# reading population data
population_data = pd.read_csv("../data/datasets/lagos_lga_population.csv", usecols=["local", "mean"])
population_data.head()
```

```python
lagos_map = folium.Map(
    location=[6.5421, 3.5079],
    zoom_start=10
)

my_map = "../data/geojson/maps/lagos_lga.geojson"
folium.Choropleth(
    geo_data=my_map,
    name='choropleth',
    data=population_data,
    columns=['local', 'mean'],
    fill_color='YlGn',
    key_on='feature.properties.NAME_2',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='LGA Population Count'
).add_to(lagos_map)

folium.LayerControl().add_to(lagos_map)

lagos_map.save("outputs/lagos_lga.html")
lagos_map
```
