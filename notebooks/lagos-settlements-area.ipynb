# Folium Lagos Settlement Area Demo

```python
import folium
```

```python
lagos_map = folium.Map(
    location=[6.5421, 3.5079],
    tiles="cartodbpositron",
    zoom_start=10
)

built_up_areas = "../data/geojson/settlements/built-up-areas.geojson"
folium.GeoJson(
    built_up_areas,
    name='builtup areas',
    style_function=lambda x: {'fillColor': '#0000ff', 'fillOpacity': 1, 'weight': 1}
).add_to(lagos_map)

hamlets = "../data/geojson/settlements/hamlets.geojson"
folium.GeoJson(
    hamlets,
    name='hamlets',
    style_function=lambda x: {'fillColor': '#00ff00', 'fillOpacity': 1, 'weight': 1}
).add_to(lagos_map)

small_settlements = "../data/geojson/settlements/small-settlements.geojson"
folium.GeoJson(
    small_settlements,
    name='small settlements',
    style_function=lambda x: {'fillColor': '#ff0000', 'fillOpacity': 1, 'weight': 1}
).add_to(lagos_map)

folium.LayerControl().add_to(lagos_map)
folium.LayerControl(collapsed=False).add_to(lagos_map)

lagos_map.save("outputs/lagos_settlement-area.html")
lagos_map
```
