# Folium Demo

```python
import folium
import pandas as pd
```

```python
# reading population data
population_data = pd.read_csv("../data/datasets/nigeria_states_population.csv", usecols=["state", "mean"])
population_data.head()
```

```python
nigeria_map = folium.Map(
    location=[9.0820, 8.6753],
    tiles="cartodbpositron",
    zoom_start=6
)

my_map = "../data/geojson/maps/nigeria_geojson.geojson"
folium.Choropleth(
    geo_data=my_map,
    name='choropleth',
    data=population_data,
    columns=['state', 'mean'],
    fill_color='OrRd',
    key_on='feature.properties.state',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Population Count'
).add_to(nigeria_map)

folium.LayerControl().add_to(nigeria_map)

nigeria_map.save("maps/nigeria.html")
nigeria_map
```
