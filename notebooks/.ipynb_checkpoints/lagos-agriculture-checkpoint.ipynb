# Folium Lagos Settlement Demo

```python
import folium
```

```python
lagos_map = folium.Map(
    location=[6.5421, 3.5079],
    zoom_start=10
)

nigeria_map = "../data/geojson/maps/lagos_geojson.geojson"
folium.GeoJson(
    nigeria_map,
    name='lagos',
    style_function=lambda x: {'fillColor': '#ffffff'}
).add_to(lagos_map)

built_up_areas = "../data/geojson/settlements/built-up-areas.geojson"
folium.GeoJson(
    built_up_areas,
    name='builtup areas',
    style_function=lambda x: {'fillColor': '#ff0000'}
).add_to(lagos_map)

folium.LayerControl().add_to(lagos_map)
folium.LayerControl(collapsed=False).add_to(lagos_map)

lagos_map.save("outputs/lagos_settlements.html")
lagos_map
```
